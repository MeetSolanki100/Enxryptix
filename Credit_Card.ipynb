```python
class0= credit_card[credit_card['Class']==0]
class1= credit_card[credit_card['Class']==1]

sampled_class0= class0.sample(n=492,random_state=42)
sampled_class1= class1.sample(n=492,random_state=42)

sampled_data= pd.concat([sampled_class0,sampled_class1],axis=0)
sampled_data= sampled_data.sample(frac=1,random_state=42).reset_index(drop=True)
sampled_data.head()
```

```python
x = sampled_data.drop(['Class','Time'],axis=1)
Y = sampled_data['Class']
```

```python
x
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X_train,X_test,Y_train,Y_test = train_test_split(x,Y,test_size=0.2,random_state=42)
```

```python
scalar = StandardScaler()
X_train = scalar.fit_transform(X_train)
X_test = scalar.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
model = LogisticRegression()
model.fit(X_train,Y_train)
```

```python
pred = model.predict(X_test)
```

```python
accuracy_score(Y_test,pred)
```

```python
pred2 = model.predict(X_train)
```

```python
accuracy_score(Y_train,pred2)
```

```python
coeffs = model.coef_[0]
coeffs
```

```python
value = np.abs(coeffs)
value
```

```python
indices = np.argsort(value)
indices = indices[::-1]
indices
```

```python
plt.figure(figsize=(10,6))
plt.bar(range(len(indices)),value[indices])
plt.xticks(range(len(indices)),x.columns[indices],rotation=90)
plt.show()
```
